```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
np.random.seed(1)
df = pd.DataFrame({
    'x1': np.random.normal(0, 2, 10000),
    'x2': np.random.normal(5, 3, 10000),
    'x3': np.random.normal(-5, 5, 10000)
})
```

```python
min_max = preprocessing.minmax_scale(df)
scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python

scaled_df = pd.DataFrame(scaled_df, columns=['x1', 'x2', 'x3'])
```

```python
scaled_df
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2,figsize = (20,6))
ax1.set_title("before scaling")
sns.kdeplot(data = df['x1'],ax=ax1)
sns.kdeplot(data = df['x2'], ax= ax1)
sns.kdeplot(data = df['x3'], ax=ax1)
ax2.set_title('after scaling')
sns.kdeplot(data= scaled_df['x1'],ax=ax2)
sns.kdeplot(data = scaled_df['x2'], ax=ax2)
sns.kdeplot(data=scaled_df['x3'], ax=ax2)
plt.show()
```

```python
min_max_df = pd.DataFrame(data = min_max, columns=['x1','x2','x3'])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2,figsize = (20,6))
ax1.set_title("before scaling")
sns.kdeplot(data = df['x1'],ax=ax1)
sns.kdeplot(data = df['x2'], ax= ax1)
sns.kdeplot(data = df['x3'], ax=ax1)
ax2.set_title('after scaling')
sns.kdeplot(data= min_max_df['x1'],ax=ax2)
sns.kdeplot(data = min_max_df['x2'], ax=ax2)
sns.kdeplot(data= min_max_df['x3'], ax=ax2)
plt.show()
```
